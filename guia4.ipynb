Despues para la columna 'attributes' , notamos que los registros estan conformados por diccionarios, y la mejor manera de ordenarlo seria sacar atributos en nuevos columnas.

```python
import json
import pandas as pd

def clean_attributes(attributes):
    try:
        if isinstance(attributes, dict):
            return attributes
        elif isinstance(attributes, str):
            return json.loads(attributes.replace("'", '"'))
        else:
            return {}
    except (ValueError, TypeError):
        return {}

df_business['attributes'] = df_business['attributes'].apply(clean_attributes)
attributes_expanded = pd.json_normalize(df_business['attributes'])

df_business = pd.concat([df_business, attributes_expanded], axis=1)
df_business.info()
```

Despues de eso, las columnas generadas se transformaran a valores booleanos para facilitar el analisis.

```python
# Definir las columnas relevantes
columns_atributes = [
    'BusinessAcceptsCreditCards', 'RestaurantsTakeOut', 'BikeParking', 'WiFi', 'RestaurantsGoodForGroups',
    'RestaurantsReservations', 'GoodForKids', 'RestaurantsDelivery',
    'HasTV','Caters', 'OutdoorSeating','DogsAllowed', 'Alcohol'
]

import json

def convert_to_boolean(value, column):
    try:
        # Si el valor es un diccionario, verificar si contiene algÃºn 'True'
        if isinstance(value, dict):
            return any(v == 'True' or v == True for v in value.values())

        elif isinstance(value, str) and value.startswith("{"):
            value_dict = json.loads(value.replace("'", '"'))
            return any(v == 'True' or v == True for v in value_dict.values())

        if column == 'WiFi':
            if value in ["'free'", "u'free'", "'paid'", "u'paid'"]:
                return True
            elif value in ["'no'", "u'no'", 'None', None]:
                return False

        if column == 'Alcohol':
            if value in ["'full_bar'", "u'full_bar'", "'beer_and_wine'", "u'beer_and_wine'"]:
                return True
            elif value in ["'none'", "u'none'", 'None', None]:
                return False

        elif value == 'True':
            return True

        return False

    except (ValueError, TypeError):
        return False

for col in columns_atributes:
    df_business[col] = df_business[col].apply(lambda x: convert_to_boolean(x, col))

df_business = df_business.drop(columns=['attributes'])

# Asegurarse de que solo se mantienen las columnas necesarias
columns_to_keep = ['business_id', 'name', 'address', 'city', 'state', 'postal_code', 'latitude', 'longitude',
                   'stars', 'review_count', 'is_open', 'categories'] + columns_atributes
df_business = df_business[columns_to_keep]

df_business.info()
```

Para la columna 'categories' definimos las 5 categorÃ­as mÃ¡s comunes y llenamos los valores nulos en 'categories' con las categorÃ­as mÃ¡s comunes aleatoriamente

```python
df_business['categories'].value_counts()
```

```python
import pandas as pd
import numpy as np

# Definir las 5 categorÃ­as mÃ¡s comunes
top_5_categories = ['Beauty & Spas, Nail Salons', 'Restaurants, Pizza', 'Nail Salons, Beauty & Spas',
                    'Pizza, Restaurants', 'Restaurants, Mexican']

n_nulls = df_business['categories'].isnull().sum()
random_categories = np.random.choice(top_5_categories, n_nulls)

# Llenar los valores nulos en 'categories' con las categorÃ­as mÃ¡s comunes aleatoriamente
df_business.loc[df_business['categories'].isnull(), 'categories'] = random_categories

print(df_business['categories'].isnull().sum())
```

```python
# Separar las categorÃ­as en listas
df_business['categories_list'] = df_business['categories'].str.split(', ')
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

mlb = MultiLabelBinarizer()

# Transformar la columna de listas de categorÃ­as en columnas binarias (dummies)
categories_dummies = mlb.fit_transform(df_business['categories_list'])

df_categories_dummies = pd.DataFrame(categories_dummies, columns=mlb.classes_)

# Unir las columnas dummies al DataFrame original
df_business = pd.concat([df_business, df_categories_dummies], axis=1)

```

```python
# Sumar las columnas dummies para obtener el nÃºmero de negocios en cada categorÃ­a
category_counts = df_categories_dummies.sum().sort_values(ascending=False)
top_100_categories = category_counts.head(100).index.tolist()
print(top_100_categories)
```

```python
def classify_category(cat_list):
    for cat in top_100_categories:
        if cat in cat_list:
            return cat
    return 'Otro'

# Crear una nueva columna 'categories_grouped' donde se clasifican las categorÃ­as
df_business['categories_grouped'] = df_business['categories'].apply(lambda x: classify_category(x.split(', ')))
```

```python
import plotly.graph_objects as go
import plotly.express as px
import pandas as pd

df_reviews['stars'] = pd.to_numeric(df_reviews['stars'], errors='coerce')

df_reviews_business = df_reviews.merge(df_business[['business_id', 'state']], on='business_id', how='left')

df_reviews_high_stars = df_reviews_business[df_reviews_business['stars'] > 4]

total_reviews_per_state = df_reviews_business['state'].value_counts().reset_index()
total_reviews_per_state.columns = ['state', 'total_reviews']

high_stars_reviews_per_state = df_reviews_high_stars['state'].value_counts().reset_index()
high_stars_reviews_per_state.columns = ['state', 'high_stars_reviews']

state_reviews = pd.merge(total_reviews_per_state, high_stars_reviews_per_state, on='state', how='left')
state_reviews['high_stars_reviews'] = state_reviews['high_stars_reviews'].fillna(0)  # Reemplazar NaN con 0
state_reviews['percentage_high_stars'] = (state_reviews['high_stars_reviews'] / state_reviews['total_reviews']) * 100

top_states = ['PA', 'FL', 'TN', 'IN', 'MO', 'LA', 'AZ', 'NJ', 'NV', 'CA', 'ID', 'DE', 'IL']
state_reviews_filtered = state_reviews[state_reviews['state'].isin(top_states)]

state_names = {
    'PA': 'Pennsylvania', 'FL': 'Florida', 'TN': 'Tennessee', 'IN': 'Indiana', 'MO': 'Missouri',
    'LA': 'Louisiana', 'AZ': 'Arizona', 'NJ': 'New Jersey', 'NV': 'Nevada', 'CA': 'California',
    'ID': 'Idaho', 'DE': 'Delaware', 'IL': 'Illinois'
}
state_reviews_filtered['state_full'] = state_reviews_filtered['state'].map(state_names)

fig = px.choropleth(
    state_reviews_filtered,
    locations='state',
    locationmode="USA-states",
    color='percentage_high_stars',
    hover_name='state_full',
    hover_data=['total_reviews', 'percentage_high_stars'],
    color_continuous_scale="Viridis",
    scope="usa",
    title="Porcentaje de reseÃ±as con calificaciÃ³n mayor a 4 por estado"
)

# AÃ±adir texto con las abreviaturas de los estados
for index, row in state_reviews_filtered.iterrows():
    fig.add_trace(go.Scattergeo(
        locationmode="USA-states",
        locations=[row['state']],
        text=row['state'],
        mode='text',
        textfont=dict(size=14, color='black'),
        hoverinfo="none",
        showlegend=False
    ))

legend_items = [f"{code}: {name}" for code, name in state_names.items()]

fig.update_layout(
    title_text="Porcentaje de reseÃ±as con calificaciÃ³n mayor a 4 por estado",
    geo_scope='usa',
    annotations=[
        go.layout.Annotation(
            text="<br>".join(["<b>Leyenda de Estados:</b>"] + legend_items),
            xref="paper", yref="paper",
            x=0.01, y=0.5, showarrow=False,
            align="left",
            font=dict(size=12, color="black")
        )
    ]
)


fig.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Definir el tamaÃ±o del chunk
chunk_size = 500000

# Seleccionar solo las columnas necesarias de df_business
df_business_reduced = df_business[['business_id']]

# Lista para guardar los resultados parciales
old_businesses_list = []
new_businesses_list = []

chunks = [df_reviews[i:i + chunk_size] for i in range(0, len(df_reviews), chunk_size)]

for chunk in chunks:

    chunk_merged = pd.merge(chunk, df_business_reduced, on='business_id', how='left')

    first_year = chunk_merged.groupby('business_id')['year'].min().reset_index()
    first_year.columns = ['business_id', 'first_review_year']

    chunk_merged = pd.merge(chunk_merged, first_year, on='business_id')

    old_businesses_chunk = chunk_merged[chunk_merged['first_review_year'] < 2014]
    new_businesses_chunk = chunk_merged[chunk_merged['first_review_year'] >= 2014]

    old_businesses_avg_chunk = old_businesses_chunk.groupby('year')['stars'].mean().reset_index()
    new_businesses_avg_chunk = new_businesses_chunk.groupby('year')['stars'].mean().reset_index()

    old_businesses_list.append(old_businesses_avg_chunk)
    new_businesses_list.append(new_businesses_avg_chunk)

old_businesses_avg = pd.concat(old_businesses_list, ignore_index=True)
new_businesses_avg = pd.concat(new_businesses_list, ignore_index=True)

old_businesses_avg = old_businesses_avg.groupby('year')['stars'].mean().reset_index()
new_businesses_avg = new_businesses_avg.groupby('year')['stars'].mean().reset_index()

plt.figure(figsize=(14, 7))

sns.lineplot(data=old_businesses_avg, x='year', y='stars', label='Negocios Antiguos', marker='o', color='blue')

sns.lineplot(data=new_businesses_avg, x='year', y='stars', label='Negocios Nuevos', marker='o', color='red')

plt.title('EvoluciÃ³n de las calificaciones promedio: Negocios Antiguos vs. Nuevos')
plt.xlabel('AÃ±o')
plt.ylabel('CalificaciÃ³n Promedio (Stars)')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Calcular el porcentaje de negocios cerrados por categorÃ­a
df_closed_business = df_business[df_business['is_open'] == 0]
df_total_business = df_business.groupby('categories_grouped').size().reset_index(name='total_business')
df_closed_count = df_closed_business.groupby('categories_grouped').size().reset_index(name='closed_business')

# Combinar los datos para calcular el porcentaje de negocios cerrados
df_closure_percentage = pd.merge(df_total_business, df_closed_count, on='categories_grouped', how='left')
df_closure_percentage['closed_business'] = df_closure_percentage['closed_business'].fillna(0)
df_closure_percentage['closure_percentage'] = (df_closure_percentage['closed_business'] / df_closure_percentage['total_business']) * 100

# Ordenar por el porcentaje de cierre
df_closure_percentage = df_closure_percentage.sort_values(by='closure_percentage', ascending=False)

plt.figure(figsize=(12, 8))
sns.barplot(x='closure_percentage', y='categories_grouped', data=df_closure_percentage, palette='coolwarm')
plt.title('Porcentaje de negocios cerrados por categorÃ­a')
plt.xlabel('Porcentaje de cierre (%)')
plt.ylabel('CategorÃ­a')
plt.show()
```
